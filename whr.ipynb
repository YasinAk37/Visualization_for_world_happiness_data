```python
#importing needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objs as go
```

```python
#importing data
df_happy = pd.read_csv("whr.csv")

df_happy.head(15)
```

```python
#heatmap
plt.figure(figsize = (12,6))
sns.heatmap(df_happy.select_dtypes("number").corr(), annot = True)
```

```python
plt.figure(figsize = (12,6))
sns.countplot(x = "year", data = df_happy)
plt.xticks(rotation = 90)
plt.show()
```

```python
# relation between Life Ladder and Generosity from 2005 to 2020

plt.figure(figsize = (14,6))
sns.scatterplot(x = "Life Ladder", y = "Generosity", hue = "year", data = df_happy)
```

```python
# relation between Life Ladder and Log GDP per capita from 2005 to 2020

plt.figure(figsize = (14,6))
sns.scatterplot(x = "Life Ladder", y = "Log GDP per capita", hue = "year", data = df_happy)
```

```python
# The most happy countries 
plt.figure(figsize = (14,6))
df_maxhappy = df_happy[df_happy.loc[:, "Life Ladder"]> 7.5]
sns.barplot(x = "Life Ladder", y ="Country name", data = df_maxhappy)
plt.show()
```

```python
# The most and the least happy countries 
plt.figure(figsize=(14, 6))
df_maxhappy = df_happy[(df_happy["Life Ladder"] > 7.5) | (df_happy["Life Ladder"] < 3.0)]
df_maxhappy_sorted = df_maxhappy.sort_values("Life Ladder", ascending=False)
sns.barplot(x="Life Ladder", y="Country name", data=df_maxhappy_sorted)
plt.show()
```

```python
#Lets find Turkiye

df_turkey = df_happy[df_happy.loc[:,"Country name"] == "Turkey"]

df_turkey
```

```python
# Lets find Turkey
plt.figure(figsize = (14,6))
df_maxhappy = df_happy[(df_happy["Life Ladder"]> 5.15) & (df_happy["Life Ladder"]< 5.2)]
sns.barplot(x = "Life Ladder", y ="Country name", data = df_maxhappy)
plt.show()
```

```python
# World map of happiness by years

harita = px.choropleth(df_happy.sort_values("year"),
                       locations = "Country name",
                       color = "Life Ladder",
                       locationmode = "country names",
                       animation_frame = "year")

harita.update_layout(title = "Happiness by year")
harita.show()
```

```python
# World map of generosity by years

harita = px.choropleth(df_happy.sort_values("year"),
                       locations = "Country name",
                       color = "Generosity",
                       locationmode = "country names",
                       animation_frame = "year")

harita.update_layout(title = "Generosity by years")
harita.show()
```

```python
# World map of Freedom to make life choices by years

harita = px.choropleth(df_happy.sort_values("year"),
                       locations = "Country name",
                       color = "Freedom to make life choices",
                       locationmode = "country names",
                       animation_frame = "year")

harita.update_layout(title = "Freedom to make life choices by years")
harita.show()
```

```python
# World map of Log GDP per capita by years

harita = px.choropleth(df_happy.sort_values("year"),
                       locations = "Country name",
                       color = "Log GDP per capita",
                       locationmode = "country names",
                       animation_frame = "year")

harita.update_layout(title = "Log GDP per capita by years")
harita.show()
```
